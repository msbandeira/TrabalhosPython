Copia do tutorial de envio de mensagem em massa pelo whatsapp #hastag Treinamentos.

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

navegador = webdriver.Chrome()
time.sleep(5)
navegador.maximize_window()
navegador.get("https://web.whatsapp.com/")

time.sleep(30)
```

```python
#importar o arquivo excel
import pandas as pd

tabela = pd.read_excel("contatos.xlsx")
display(tabela[['Telefone', 'Mensagem', 'Arquivo']])
```

```python
#loop para envio das mensagens

import urllib
import time
import os

for linha in tabela.index:
    
    mensagem = tabela.loc[linha, "Mensagem"]
    telefone = tabela.loc[linha, "Telefone"]
    arquivo = tabela.loc[linha, "Arquivo"]    

    #enviar mensagens
    link = f"https://web.whatsapp.com/send?phone={telefone}&text={mensagem}"

    navegador.get(link)
    #esperar a tela do whatsapp carregar com a mensagem
    time.sleep(10)
    
    navegador.find_element(By.XPATH, '//*[@id="main"]/footer/div[1]/div/span[2]/div/div[2]/div[2]/button/span').click()
    time.sleep(5)
    
    #anexos
    if arquivo != 0:
        caminho_completo = os.path.abspath(arquivo)
        navegador.find_element(By.XPATH, '//*[@id="main"]/footer/div[1]/div/span[2]/div/div[1]/div[2]/div/div/div/span').click()
        time.sleep(2)
        navegador.find_element(By.XPATH, '//*[@id="main"]/footer/div[1]/div/span[2]/div/div[1]/div[2]/div/span/div/ul/div/div[2]/li/div/input').send_keys(caminho_completo)
        time.sleep(5)
        navegador.find_element(By.XPATH, '//*[@id="app"]/div/div[2]/div[2]/div[2]/span/div/div/div/div[2]/div/div[2]/div[2]/div/div').click()
        time.sleep(10)

    
```

```python
navegador.find_element(By.XPATH, '//*[@id="app"]/div/div[2]/div[3]/header/div[2]/div/span/div[5]/div/span').click()
time.sleep(3)
navegador.find_element(By.XPATH, '//*[@id="app"]/div/div[2]/div[3]/header/div[2]/div/span/div[5]/span/div/ul/li[8]/div').click()
time.sleep(3)
navegador.find_element(By.XPATH, '//*[@id="app"]/div/span[2]/div/div/div/div/div/div/div[3]/div/button[2]/div/div').click()
navegador.close()
```
